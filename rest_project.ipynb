```sql
UPDATE rest_final
SET time_of_day = CASE 
    WHEN hour >= 5  AND hour < 12 THEN 'Morning'
    WHEN hour >= 12 AND hour < 17 THEN 'Afternoon'
    WHEN hour >= 17 AND hour < 21 THEN 'Evening'
    ELSE 'Night' 
END;
```

```sql
UPDATE rest_final
SET is_weekend = CASE 
    WHEN date_format(order_date, 'EEEE') IN ('Friday', 'Saturday') THEN 1 
    ELSE 0 
END;
```

```sql
SELECT 
    branch,
    COUNT(CASE WHEN total_amount < 0 THEN 1 END) AS refund_count,
    COUNT(CASE WHEN total_amount > 0 THEN 1 END) AS sales_count,
    ABS(SUM(CASE WHEN total_amount < 0 THEN total_amount ELSE 0 END)) AS total_refund_value,
    SUM(CASE WHEN total_amount > 0 THEN total_amount ELSE 0 END) AS gross_revenue,
    (COUNT(CASE WHEN total_amount < 0 THEN 1 END) / COUNT(*)) * 100 AS refund_rate_percentage
FROM rest_final
GROUP BY branch
ORDER BY refund_rate_percentage DESC;
```

```sql
SELECT 
    item_name,
    branch,
    MIN(price) AS Min_Price,
    MAX(price) AS Max_Price,
    ROUND(AVG(price), 2) AS Avg_Price,
    COUNT(*) AS Total_Orders
FROM rest_final
GROUP BY item_name, branch
ORDER BY item_name ASC, Min_Price DESC;
```

```sql
SELECT 
    customer_id,
    AVG(total_amount) AS avg_spend,
    COUNT(order_id) AS total_visits,
    CASE 
        WHEN AVG(total_amount) < 150 THEN 'Tier 1 (Budget)'
        WHEN AVG(total_amount) < 500 THEN 'Tier 2 (Mid-Range)'
        ELSE 'VIP (High Spender)' 
    END AS customer_tier
FROM rest_final
GROUP BY customer_id;
```
